```python
#importing datasets
data_set= pd.read_csv('100 Records.csv')
```

```python
#Extracting Independent and dependent Variable
x= data_set.iloc[:, [2,3]].values
y= data_set.iloc[:, 4].values
```

```python
# Splitting the dataset into training and test set.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=0)
```

```python
#feature Scaling
#StandardScaler function Normalises the data
from sklearn.preprocessing import StandardScaler
st_x= StandardScaler()
x_train= st_x.fit_transform(x_train)
x_test= st_x.transform(x_test)
#fit ka kaam hota hai normalization mei sigma amd mean calculate krna
#transform ka kaam hota hai x-u/sigma ki values ko calculate krna
```

```python
#Fitting Logistic Regression to the training set
from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression()
classifier.fit(x_train, y_train)
#Trainig krne k liye we use fit
```

```python
#Predicting the test set result
y_pred= classifier.predict(x_test)
```

```python
mlr_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred})
mlr_diff
```

```python
#Creating the Confusion matrix
#confusion matrix is used to check the performance of the classifier
from sklearn.metrics import confusion_matrix , ConfusionMatrixDisplay
cm= confusion_matrix(y_test, y_pred)
cm
```

```python
ConfusionMatrixDisplay(cm).plot()
```
